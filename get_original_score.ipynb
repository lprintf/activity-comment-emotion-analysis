# 数据预处理

1. 根据商品活动时间筛选商品，得到以商品为索引的，某时间段的表格。
2. 以商品评论为索引组织数据，得到列数为 商品数*平均评论数 的表格。

```python
import pandas as pd
from dateutil.parser import parse
```

```python
activity_path = "./data/activity.xlsx"
activity_sheet = pd.read_excel(activity_path)
activity_sheet.head()
```

```python
allcomment_path = "./data/comment3.csv"
comment_sheet = pd.read_csv(allcomment_path)
comment_sheet.head()
```

```python
comment_activity_sheet = pd.merge(
    comment_sheet,
    activity_sheet[["sku_id", "活动时间"]],
    left_on="productId",
    right_on="sku_id",
)
comment_activity_sheet.head()
```

```python
july = "2020-07-01 00:00:00"
august = "2020-08-01 00:00:00"
september = "2020-09-01 00:00:00"
```

```python
after_february = comment_activity_sheet[
    comment_activity_sheet["活动时间"]
    > "2020-02-01 00:00:00"
]

before_july = after_february[
    after_february["活动时间"] < july
]
before_august = after_february[
    after_february["活动时间"] < august
]
before_september = after_february[
    after_february["活动时间"] < september
]
```

```python
'before_july: ',before_july.shape,\
'before_august: ',before_august.shape,\
'before_september: ',before_september.shape,\
```

```python
import datetime

max_time_span = datetime.timedelta(days=90)
zero_time_span = datetime.timedelta(days=0)
```

```python
may='2020-05-01 00:00:00'
november='2020-11-01 00:00:00'

def get_comments(x):
    result=[]
    
    time_span = datetime.timedelta(days=0)
    for y in eval(x[1]).values():
        for z in y:
            _span = x[0]-parse(z["referenceTime"])
            if _span > zero_time_span:
                if _span > max_time_span:
                    time_span=max_time_span
                    break
                elif _span>time_span:
                    time_span=_span
                    


    time_ceiling=x[0]+time_span
    time_floor=x[0]-time_span
    for y in eval(x[1]).values():
        for z in y:
            if parse(z["referenceTime"])>time_floor and parse(z["referenceTime"])<time_ceiling:
                result.append((z["referenceTime"], z["score"], z["content"]))
    return result

# before_august.dropna(subset=['product_comments'])
product_comment=before_august[['活动时间','product_comments']].apply(get_comments,axis=1)
```

```python
def get_comments_(x):
    result=[]

        
    time_span = datetime.timedelta(days=0)
    for y in eval(x[1]):
        _span = x[0]-parse(y["referenceTime"])
        if _span > zero_time_span:
            if _span > max_time_span:
                time_span=max_time_span
                break
            elif _span>time_span:
                time_span=_span

    time_ceiling=x[0]+time_span
    time_floor=x[0]-time_span
    for y in eval(x[1]):
        if parse(y["referenceTime"])>time_floor and parse(y["referenceTime"])<time_ceiling:
            result.append((y["referenceTime"], y["score"], y["content"]))
    return result
    
# todo 可合并
poor_comment=before_august[['活动时间','poor_comment']].apply(get_comments_,axis=1)
middle_comment=before_august[['活动时间','middle_comment']].apply(get_comments_,axis=1)
good_comment=before_august[['活动时间','good_comment']].apply(get_comments_,axis=1)
```

```python
comment=poor_comment+middle_comment+good_comment+product_comment
comment.values[0]
```

```python
# comment.map(lambda x: list(set(x)))
def deduplicate(x:list):
    x_=list(set(x))
    print(len(x),' to ',len(x_))
    return x_

deduplicate_comment=comment.map(deduplicate)
```

```python
before_august["评论明细"] = deduplicate_comment
# before_august[['productId', '评论明细']].values[0]
```

```python
thin_before_august = pd.merge(before_august[['productId','评论明细']], activity_sheet[['sku_id', '活动时间']], left_on='productId', right_on='sku_id')


thin_before_august['活动前评论'] = thin_before_august.apply(lambda x: [c for c in x['评论明细'] if parse(c[0]) < x['活动时间']], axis=1)
thin_before_august['活动后评论'] = thin_before_august.apply(lambda x: [c for c in x['评论明细'] if parse(c[0]) > x['活动时间']], axis=1)

thin_before_august.head()
```

```python
data4train=[]
```

```python
for value in thin_before_august.values:
    for __comment in value[4]:
        data4train.append([value[0], value[3], __comment[0], __comment[2], '活动前', __comment[1]])
    for __comment in value[5]:
        data4train.append([value[0], value[3], __comment[0], __comment[2], '活动后', __comment[1]])
```

```python
data4train_sheet = pd.DataFrame(data4train, columns=['productId', '活动时间', '评论时间', '评论内容', '类型', '评分'])
# data4train_sheet['美学维度'] = ''
# data4train_sheet['功能维度'] = ''
# data4train_sheet['服务维度'] = ''
```
